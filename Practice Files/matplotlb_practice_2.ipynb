```python
#Import Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(y_axis, x_axis,'o', c='red', label="Chicago")
plt.title('PyBer Fare by Month')
plt.xlabel("Month")
plt.ylabel("Fare ($)")
plt.legend()
plt.xlim(0, 50)
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis, c='blue')
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#List comprehension
#new_list = [expression for item in list if coinditional]
plt.scatter(x_axis, y_axis, s = [i*3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*5)
ax.scatter(y_axis, x_axis, s=y_axis_larger, c='skyblue', alpha=0.8, edgecolors='black', linewidth=2, label='Boston')
ax.set_title('PyBerFare by Month')
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
ax.legend()
ax.set_xlim(0, 50)
ax.invert_yaxis()

```
